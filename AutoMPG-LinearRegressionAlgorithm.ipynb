### 2. Load and process data
1. In the next cell I am importing the data from the file into Pands data frame,
2. I have noticed some invalid or incomplete data, so I am pruning the data by removing NaN and invalid data.
3. I am randomly splitting data into 80% train and 20% test sets
4. Then I am converting the data into numpy arrays as follows

```python
def load_data():
    # 1. Importing the data from the csv file downloaded from https://archive.ics.uci.edu/ml/machine-learning-databases/auto-mpg/
    df = pd.read_csv("auto-mpg.data",delim_whitespace=True,index_col=8,names=['mpg','cylinders','displacement','bhp','weight','acc','year','orgin','name'])
    #print('Data before pruning: '+str(df.shape))
    # 2. removing null NaN values from dataframe
    df = df.dropna(axis=0, how='any')
    # 2. removing invalid data which have '?'
    df = df[(df.bhp != '?')]
    #print('Data after pruning: '+str(df.shape))
    #print(df[:5])
    df['bhp'] = df.bhp.astype(float)
    df['cylinders'] = df.cylinders.astype(float)
    df['year'] = df.year.astype(float)
    df['orgin'] = df.orgin.astype(float)
    #print(df.dtypes)
    # Ex3 splitting data into test and train randomly
    np.random.seed(0)
    msk = np.random.rand(len(df)) < 0.8

    df_train = df[msk]
    df_test = df[~msk]

    # Now we have two different data frames for train and test
    # Now we will seprate the target from data frame

    y_train = df_train[['mpg']].values
    x_train = df_train[['cylinders','displacement','bhp','weight','acc','year', 'orgin']].values

    y_test = df_test[['mpg']].values
    x_test = df_test[['cylinders','displacement','bhp','weight','acc','year', 'orgin']].values


    #print('xtrain -- > '+str(x_train.shape)+', ytrain --> '+str(y_train.shape))
    #print('xtest  -- > '+str(x_test.shape)+', ytest --> '+str(y_test.shape))
    
    return x_train, y_train, x_test, y_test, df, df_train, df_test
```

### 4 Functions for Linear Regression
We will create the following functions

1. initialize_weights - It will initialize the W and b to zeros depending on the number of variables
2. normalize_data - it will scale the values in the range 0-1 by dividing the average of each feature with standard deviation
3. linear_activation - it will do the linear regression to compute yhat h(x).
4. compute_cost - it will compute the cost of the linear regression
5. compute_derivatives - it will compute the derivatis of W and b i.e. dW, db
6. gradient_decent - it will use above funcitos to perform gradiant decent

```python
# 1. Initialize weights
def initialize_weights(X):
    """
    This function takes 2 dimensional np array X as input which has the shape n * m 
    where n -> number of rows of data, m -> number of fetures
    returns W -> m * 1 dimensional np array with all the values as 0
    b -> number with zero
    """
    W = np.zeros((X.shape[1],1))
    b = 0
    return W, b
```

```python
# 2. normalize_data
def normalize_data(X):
    """
    This function will normalize the data to get them into the range between 0 an 1
    The X_norm = (X - mean)/Standard deviation
    """
    x_mean = np.mean(X, axis = 0, keepdims = True)
    x_std = np.std(X, axis = 0, keepdims = True)
    x_diff = X - x_mean
    X_norm = x_diff/x_std
    return X_norm
```

```python
# Calculate the h(x) or Y hat, I vectorizing by using numpy libs
def linear_function(X, W, b):
    yhat = (np.dot(X,W)) + b
    return yhat;
```

```python
# This is the vectorized implementation for computing the compute_error_for_line_given_points
def compute_cost(yhat,Y):
    m = yhat.shape[0]
    J = (1.0/(2*m))*np.sum((np.power((yhat - Y),2)))
    return J
```

```python
# Testing compute_cost
yh = np.array([[3],[3]])
y = np.array([[1],[1]])
j= compute_cost(yh,y)
print(j)
```

```python
# This function will calculate the partial derivatives of the Weights and b and multiplies it with the learnong rate
def compute_derivatives(Y, Yhat, X, alpha):
    m = Yhat.shape[0]
    diff = Yhat - Y
    db = (alpha/m) * (np.sum(diff))
    dw = (alpha/m) * (np.sum((X * diff), axis = 0, keepdims = True))
    return dw.T, db
```

```python
# Testing compute_derivatives
yht = np.array([[2],[2]])
yt = np.array([[1],[1]])
xt = np.array([[1, 1, 1, 1],[2, 2, 2, 2]])
wt, bt = initialize_weights(xt)

print(compute_derivatives(yt, yht, xt, 0.5))
```

```python
# step_gradient and gradient_descent_runner functions merged together
def gradient_decent(X, Y, learning_rate, num_iterations):
    W, b = initialize_weights(X)
    costs = []
    for i in range(num_iterations):
        yhat = linear_function(X, W, b)
        cost = compute_cost(yhat,Y)
        costs.append(cost)
        dw, db = compute_derivatives(Y, yhat, X, learning_rate)
        W = W - dw
        b = b - db
    return costs, W, b
```

### 5 Train with different learning rated
In the following block I called the gradient decent with different learning rates

```python
# Trying the gradient decent with different learing rates

# loading the data set
x_train, y_train, x_test, y_test, df, df_train, df_test = load_data()

# normalizing the traing set
x_train = normalize_data(x_train)

# normalizing the test set
x_test = normalize_data(x_test)

# intitializing different learning rates
learning_rates = [0.0001,0.0005,0.001,0.005,0.01,0.05,0.1,0.5,1]

# seting the number of iterations
num_iterations = 5000

# setting the figure size and shape
figure(num=None, figsize=(12, 12), dpi=80, facecolor='w', edgecolor='k')
i = 330

# initializing the dictionary to store the weights for different lrs
weights = {}

# looping and training for different LRs
for lr in learning_rates :
    costs, W, b = gradient_decent(x_train,y_train,lr,num_iterations)
    
    # saving the weights and bias for a lr
    weights['W_' + str(lr)] = W
    weights['b_' + str(lr)] = b
    
    # plotting the costs over num iteration
    i += 1
    plt.figure(1)
    plt.subplot(i)
    plt.plot(costs, color='blue',linewidth=3.0)
    plt.title('Plot with learning rate : ' + str(lr))
    print("The final cost with learning rate %s after %s iterations is %s" % (lr,num_iterations,costs[-1]))
    
```

From the above plots we can see that as we increase the learning rate the cost is decresing faster but after a thasholed the cost is increasing instead, So the ideal learing rate would be 0.1, below code will get the final weights with learing rate 0.1

```python
# The coefficients
W = weights['W_0.1']
b = weights['b_0.1']
print('Final weights with learing rate 0.1 are : W : '+ str(W) + ', b : ' + str(b))
```

### RMSE function to calculate the errors
Now we will implement a function to calculate the Root mean squared error to get the error rate of the predicted MPG vs actual MPG

```python
#RMSE function
def rmse(predictions, actual):
    return np.sqrt(((predictions - actual) ** 2).mean())
    
```

### Predit using test set

We will now predic the MPGs from the test set and and the training set and calculate the RMSE on both test and training sets.
We found the RSME to be 3.54 that means the predicted MPD is + or - 3.54 mpg

```python
# Make predictions using the testing set
y_test_pred = linear_function(x_test, W, b)
y_train_pred = linear_function(x_train, W, b)
train_rmse = rmse(y_train_pred, y_train)
test_rmse = rmse(y_test_pred, y_test)
print ('RMSE Train : '+str(train_rmse))
print ('RMSE Test : '+str(test_rmse))

print('Y pred : '+str(y_test_pred[:10,:].round(2)))
print('Y actu : '+str(y_test[:10,:]))
```
